```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import export_text
```

```python
data = pd.read_csv('mushroom1.csv')
```

```python
columns_to_analyze = [col for col in data.columns if col != 'class']
for col in columns_to_analyze:
    percentage_table = data.groupby(['class', col]).size().unstack(fill_value=0)
    percentage_table = percentage_table.apply(lambda x: x / x.sum() * 100, axis=1)

    # Heatmap
    sns.heatmap(percentage_table, annot=True, fmt=".1f", cmap="coolwarm")
    plt.title(f"Percentage Distribution of {col} by Class")
    plt.show()
```

This code visualizes the percentage distribution of each feature (excluding the 'class' column) with respect to the 'class' column using heatmaps. Heres a step-by-step breakdown:

1. **Excludes the 'class' Column**: The `columns_to_analyze` list contains all the columns from the `data` DataFrame except the 'class' column, which is assumed to be the target variable.

2. **Calculates Percentage Distribution**: For each column in `columns_to_analyze`, the code:
   - Groups the data by the 'class' and the current column (`col`) using `groupby(['class', col])` and counts the occurrences of each combination with `.size()`. This creates a frequency table.
   - Uses `.unstack(fill_value=0)` to reshape the table, filling any missing values with `0`.
   - Converts the count values to percentages by dividing each value by the sum of its corresponding class (across rows), and multiplying by 100 to get the percentage.

3. **Generates a Heatmap**: For each column:
   - A heatmap is generated using `sns.heatmap()`, where:
     - `annot=True` displays the numerical values on the heatmap.
     - `fmt=".1f"` formats the annotations to one decimal place.
     - `cmap="coolwarm"` sets the color scheme of the heatmap.
   - The heatmap is titled with the column name to describe the visualization.
   - `plt.show()` is called to display the heatmap.

This code provides a visual representation of how the distribution of each feature (in terms of percentages) varies across different classes. The heatmap allows easy identification of trends, patterns, or imbalances between the features and the target variable.

```python
# Loop through each column and create bar plots
for col in columns_to_analyze:
    print(f"Generating bar plot for column: {col}")

    # Calculate percentage distribution
    percentage_table = data.groupby(['class', col]).size().unstack(fill_value=0)
    percentage_table = percentage_table.apply(lambda x: x / x.sum() * 100, axis=1)

    # Create a bar plot
    percentage_table.T.plot(kind='bar', figsize=(10, 6), edgecolor='k')

    # Customizing the plot
    plt.title(f"Percentage Distribution of {col} by Class", fontsize=14)
    plt.xlabel(col, fontsize=12)
    plt.ylabel("Percentage (%)", fontsize=12)
    plt.xticks(rotation=45, ha='right', fontsize=10)
    plt.legend(title="Class", fontsize=10)
    plt.tight_layout()

    # Show the plot
    plt.show()
```

This code generates bar plots to visualize the percentage distribution of each feature (excluding the 'class' column) with respect to the target variable ('class'). Heres how it works step by step:

1. **Excludes the 'class' Column**: The list `columns_to_analyze` contains all columns from the `data` DataFrame except for the 'class' column, which is the target variable.

2. **Calculates Percentage Distribution**: For each column in `columns_to_analyze`, the code:
   - Groups the data by both the 'class' and the current column (`col`) using `groupby(['class', col])`, then counts the occurrences of each combination with `.size()`.
   - The result is reshaped with `.unstack(fill_value=0)` to create a table where rows represent the 'class' and columns represent the unique values of the current feature.
   - It then normalizes the counts to percentages by dividing each value by the sum of its corresponding class (across rows) and multiplying by 100.

3. **Generates Bar Plots**: For each column:
   - The transposed percentage table (`percentage_table.T`) is plotted as a bar plot using `plot(kind='bar')`. The `figsize=(10, 6)` parameter defines the size of the plot, and `edgecolor='k'` adds black edges to the bars for better visibility.
   - Several customizations are applied:
     - A title is added with `plt.title()`, indicating the column being analyzed.
     - The x-axis label (`plt.xlabel()`) represents the feature being plotted, and the y-axis (`plt.ylabel()`) shows the percentage values.
     - `plt.xticks(rotation=45, ha='right')` rotates the x-axis labels for better readability, and `fontsize=10` adjusts their size.
     - A legend is included with `plt.legend()`, which is titled "Class" to distinguish the class labels.
     - `plt.tight_layout()` ensures that the plot is displayed without clipping or overlap.

4. **Displays the Plot**: Finally, `plt.show()` is called to display the bar plot.

This code provides a clear visual representation of the percentage distribution of each feature across different classes, making it easier to interpret the relationship between features and the target variable.
